```python
pathname = "./data/gr_user/*.vtk"
files = glob.glob(pathname)
```

```python
def get_data(name):
    filenames = glob.glob("./data/{name}/*.vtk".format(name=name))
    indices = [int(f[-9:-4]) for f in filenames]
    sorted_names = [f[1] for f in sorted(zip(indices,filenames), key = lambda x: x[0])]
    Nt = len(indices)
    
    x, y, z, data0 = athena_read.vtk(sorted_names[0])
    
    
    B_shape = list(data0["Bcc"].shape)
    B_shape.append(Nt)
    B = np.zeros(B_shape[1:5])
    B[:,:,:,0] = data0["Bcc"][0,:,:,:]
    
    for i in range(1, Nt):
        x, y, z, datai = athena_read.vtk(sorted_names[i])
        B[:,:,:,i] = datai["Bcc"][0,:,:,:]
    
    
    return B
```
